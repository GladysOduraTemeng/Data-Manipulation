### Description
This dataset captures user-generated text along with its corresponding sentiment label and a recommended song. The songs are categorized based on various attributes such as genre, tempo (BPM), mood, energy, and danceability. 

### Objective
The primary aim of this dataset is to analyze how user sentiment influences music recommendations and to develop a system that suggests songs tailored to users' emotional states.

### Business Questions
- How effectively does sentiment analysis help in recommending music that aligns with users' emotions?
- Which music attributes (e.g., genre, BPM, mood, energy, danceability) are most closely associated with specific sentiments?
- Are there patterns in user sentiment that indicate a preference for specific types of music?
- Can this approach be scaled to improve real-time music recommendation systems on streaming platforms?
- How do different genres contribute to emotional regulation based on user sentiment?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
music = pd.read_csv('music_sentiment_dataset.csv')

music
```

#### Handling of missing and duplicates values

```python
#Checking of duplicated values
music.duplicated().sum()
```

```python
#Checking of missing values
music.isna().sum()
```

```python
# Convert numerical columns to integer type
music["Tempo (BPM)"] = music["Tempo (BPM)"].astype(int)
```

```python
# Set global plot style
plt.style.use("ggplot")
```

```python
# 1. Sentiment Distribution
plt.figure(figsize=(6, 4))
music["Sentiment_Label"].value_counts().plot(kind="bar", color=["blue", "green", "orange","yellow"])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
# 2. Tempo Distribution by Sentiment
plt.figure(figsize=(6, 4))
music.boxplot(column="Tempo (BPM)", by="Sentiment_Label", grid=False, patch_artist=True)
plt.title("Tempo (BPM) by Sentiment")
# plt.suptitle("")  # Remove automatic title
plt.xlabel("Sentiment")
plt.ylabel("Tempo (BPM)")
plt.show()
```

```python
# 3. Energy Levels by Sentiment (Stacked Bar Chart)
energy_counts = music.groupby(["Sentiment_Label", "Energy"]).size().unstack()
energy_counts.plot(kind="bar", stacked=True, figsize=(6, 4), color=["red", "green"])
plt.title("Energy Levels by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.legend(title="Energy")
plt.xticks(rotation=0)
plt.show()
```

```python
# 4. Danceability by Sentiment (Stacked Bar Chart)
dance_counts = music.groupby(["Sentiment_Label", "Danceability"]).size().unstack()
dance_counts.plot(kind="bar", stacked=True, figsize=(8, 6), color=["purple", "yellow"])
plt.title("Danceability by Sentiment", fontweight='bold')
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.legend(title="Danceability")
plt.xticks(rotation=0)
plt.show()
```
